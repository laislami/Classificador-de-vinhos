```python
#O algoritmo irá classificar os vinhos entre branco e tinto
#Os dados serão divididos para treino e para teste
```

```python
#Importar bibliotecas
import pandas as pd
```

```python
#Importar dados
wineDs = pd.read_csv('C:/Users/Laisla Francisco/Documents/Python/wine_dataset.csv')
```

```python
#Visualizar 5 primeiras linhas do arquivo
wineDs.head()
```

```python
#Transformar coluna de branco/tinto em valor numérico
#tinto = 0 e branco = 1
wineDs['style'] = wineDs['style'].replace('red', 0)
wineDs['style'] = wineDs['style'].replace('white', 1)
wineDs.head()
```

```python
#Separar variáveis entre preditoras e variável alvo ( style = branco/tinto)
#y armazena a coluna style
y = wineDs['style']
#x armazena todas colunas exceto a style
x = wineDs.drop('style', axis = 1)
```

```python
#Separar dados para treino e para teste
from sklearn.model_selection import train_test_split

#Irá utilizar 30% dos dados para teste e os 70% restantes de treino
#4 saídas nas variáveis de treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size = 0.3)

#Visualizar qtd de dados de treino e teste
print(x_treino.shape)
print(x_teste.shape)
```

```python
#Importar algoritmo de arvore de decisão
#Roda em cima dos dados de treino
from sklearn.ensemble import ExtraTreesClassifier

#A variável modelo recebe o algoritmo
modelo = ExtraTreesClassifier()
#O algoritmo é aplicado aos dados de treino
modelo.fit(x_treino, y_treino)

#Resultados
#Os dados foram relacionados no treino
#A função score compara os dados reais de teste, passa pro algoritmo
#Ele tenta prever a classe de cada amostra
#Exibe o quanto o algoritmo acertou dos dados inseridos
resultado = modelo.score(x_teste, y_teste)
print('Acurácia: ', resultado)
```
